```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from matplotlib.gridspec import GridSpec
import matplotlib.font_manager as fm

# 日本語フォントの設定（Windowsの場合）
# plt.rcParams['font.family'] = 'MS Gothic'
# Macの場合
# plt.rcParams['font.family'] = 'AppleGothic'
# もしくはインストールされている日本語フォントを指定

# フォントパスを追加（プラットフォームに合わせて変更）
# font_path = '/System/Library/Fonts/ヒラギノ角ゴシック W3.ttc'  # Macの例
# prop = fm.FontProperties(fname=font_path)
# プロパティなしの場合はコメントアウトして、下記のpropをすべて削除

# グラフの全体サイズを設定
plt.figure(figsize=(16, 12))
gs = GridSpec(2, 2, height_ratios=[1, 1], width_ratios=[1, 1], hspace=0.4, wspace=0.3)

# 1. 正規分布
ax1 = plt.subplot(gs[0, 0])
x = np.linspace(-5, 5, 1000)
for mu, sigma, color, label in zip([0, 0, 0], [0.5, 1, 2], ['#3366CC', '#66A3FF', '#99CCFF'], ['σ=0.5', 'σ=1', 'σ=2']):
    y = stats.norm.pdf(x, mu, sigma)
    ax1.plot(x, y, color=color, linewidth=2, label=label)
ax1.set_xlim(-5, 5)
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.set_title('正規分布 (Normal Distribution)', fontsize=14)  # prop=propを追加可能
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('確率密度 f(x)', fontsize=12)  # prop=propを追加可能
ax1.legend()

# 2. 二項分布
ax2 = plt.subplot(gs[0, 1])
n_values = [10, 20, 40]
p = 0.5
x = np.arange(0, max(n_values)+1, 1)
for n, color, linestyle in zip(n_values, ['#FF9900', '#FFB84D', '#FFCC80'], ['-', '--', ':']):
    y = stats.binom.pmf(x, n, p)
    ax2.bar(x[:n+1], y[:n+1], alpha=0.6, color=color, label=f'n={n}, p={p}', width=0.8)
    ax2.plot(x[:n+1], y[:n+1], color='black', linestyle=linestyle, linewidth=1)
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.set_title('二項分布 (Binomial Distribution)', fontsize=14)  # prop=propを追加可能
ax2.set_xlabel('成功回数 k', fontsize=12)  # prop=propを追加可能
ax2.set_ylabel('確率 P(X=k)', fontsize=12)  # prop=propを追加可能
ax2.legend()

# 3. ポアソン分布
ax3 = plt.subplot(gs[1, 0])
x = np.arange(0, 20, 1)
for lambda_, color, marker in zip([1, 3, 7], ['#33CC33', '#66E066', '#99EE99'], ['o', 's', '^']):
    y = stats.poisson.pmf(x, lambda_)
    ax3.plot(x, y, color=color, marker=marker, linestyle='-', linewidth=2,
             markersize=6, label=f'λ={lambda_}')
ax3.grid(True, linestyle='--', alpha=0.7)
ax3.set_title('ポアソン分布 (Poisson Distribution)', fontsize=14)  # prop=propを追加可能
ax3.set_xlabel('発生回数 k', fontsize=12)  # prop=propを追加可能
ax3.set_ylabel('確率 P(X=k)', fontsize=12)  # prop=propを追加可能
ax3.legend()

# 4. 指数分布
ax4 = plt.subplot(gs[1, 1])
x = np.linspace(0, 5, 1000)
for lambda_, color, linestyle in zip([0.5, 1, 2], ['#CC33CC', '#D466D4', '#E699E6'], ['-', '--', ':']):
    y = stats.expon.pdf(x, scale=1/lambda_)
    ax4.plot(x, y, color=color, linestyle=linestyle, linewidth=2, label=f'λ={lambda_}')
ax4.grid(True, linestyle='--', alpha=0.7)
ax4.set_title('指数分布 (Exponential Distribution)', fontsize=14)  # prop=propを追加可能
ax4.set_xlabel('時間 x', fontsize=12)  # prop=propを追加可能
ax4.set_ylabel('確率密度 f(x)', fontsize=12)  # prop=propを追加可能
ax4.legend()

# メインタイトル
plt.suptitle('主な確率分布のグラフ', fontsize=20)  # prop=propを追加可能

# レイアウトの調整
plt.tight_layout(rect=[0, 0, 1, 0.95])

# 保存と表示
plt.savefig('probability_distributions.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, Polygon, Ellipse
from matplotlib.patches import ConnectionPatch
from matplotlib import colors
import matplotlib.patheffects as PathEffects
import matplotlib.font_manager as fm

# 日本語フォントの設定
# 環境に合わせてコメントアウトを解除してください
# Mac OSの場合
# plt.rcParams['font.family'] = 'AppleGothic'
# Windowsの場合
# plt.rcParams['font.family'] = 'MS Gothic'
# または
# font_path = '/System/Library/Fonts/ヒラギノ角ゴシック W3.ttc'  # Macの例
# prop = fm.FontProperties(fname=font_path)

# 図の設定
fig, ax = plt.subplots(figsize=(10, 8))
ax.set_aspect('equal')
ax.set_xlim(-5, 5)
ax.set_ylim(-4, 4)
ax.axis('off')

# 色の設定
bg_color = '#f8f9fa'
a_color = '#b3d9ff'
b_color = '#ffcccc'
ab_color = '#e6b3cc'
text_color_a = '#3366CC'
text_color_b = '#CC3333'
text_color_ab = '#993366'

# 全体の四角形（全標本空間Ω）
rect = Rectangle((-4.5, -3.5), 9, 7, fill=True, color=bg_color,
                 edgecolor='gray', linewidth=2, alpha=1)
ax.add_patch(rect)
ax.text(3.6, -3.1, '全事象', fontsize=14, color='gray')

# 事象A（左の楕円）
a = Ellipse((-2, 0), 4, 5, fill=True, alpha=0.7, color=a_color,
            edgecolor=text_color_a, linewidth=2)
ax.add_patch(a)

# 事象B（右の楕円）
b = Ellipse((2, 0), 4, 5, fill=True, alpha=0.7, color=b_color,
            edgecolor=text_color_b, linewidth=2)
ax.add_patch(b)

# 共通部分（A∩B）- 楕円が重なる部分は自動的に色が混ざります
# 明示的に強調するため、別のパッチを追加
intersection = Ellipse((0, 0), 2, 3.5, fill=True, alpha=0.8, color=ab_color,
                      edgecolor=text_color_ab, linewidth=2, linestyle='-')
ax.add_patch(intersection)

# ラベルを追加
# 事象A
ax.text(-3, 2.5, '事象A', fontsize=18, fontweight='bold', color=text_color_a)
ax.text(-3, 2, '（原因）', fontsize=14, color=text_color_a)
ax.text(-3, 1.5, 'P(A)', fontsize=14, color=text_color_a)

# 事象B
ax.text(2.5, 2.5, '事象B', fontsize=18, fontweight='bold', color=text_color_b)
ax.text(2.5, 2, '（結果）', fontsize=14, color=text_color_b)
ax.text(2.5, 1.5, 'P(B)', fontsize=14, color=text_color_b)

# 共通部分（A∩B）
ax.text(0, 0.2, 'A∩B', fontsize=16, fontweight='bold', color=text_color_ab,
        ha='center')
ax.text(0, -0.3, 'P(A∩B)', fontsize=14, color=text_color_ab, ha='center')

# ベイズの定理の数式
title_text = ax.text(0, 3.8, 'ベイズの定理の図解', fontsize=22, fontweight='bold',
                    ha='center')
formula = r'$P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}$'
formula_text = ax.text(0, 3, formula, fontsize=18, ha='center')

# 説明テキスト
def add_text_with_outline(x, y, text, color, size=14, outline_color='white', outline_width=3):
    text_obj = ax.text(x, y, text, fontsize=size, color=color)
    text_obj.set_path_effects([PathEffects.withStroke(linewidth=outline_width, foreground=outline_color)])
    return text_obj

# 説明テキストを追加
y_pos = -4.5
texts = [
    ('P(A) : 事前確率（事象Aが起こる確率）', text_color_a),
    ('P(B) : 周辺尤度（事象Bが起こる確率）', text_color_b),
    ('P(B|A) : 尤度（AのときにBが起こる条件付き確率）', text_color_ab),
    ('P(A|B) : 事後確率（Bが起きたときにAである確率）', text_color_ab)
]

# テキストを図の下部に配置
for i, (txt, color) in enumerate(texts):
    ax.text(-4.4, -2.5 - i*0.5, txt, fontsize=14, color=color)

# 矢印と補助説明
# 条件付き確率 P(B|A) の説明（AからBへの矢印）
arrow_connection = ConnectionPatch((-1.5, 1), (1.5, 1), 'data', 'data',
                                  arrowstyle='->', mutation_scale=15,
                                  linewidth=2, color=text_color_ab)
ax.add_patch(arrow_connection)
ax.text(0, 1.3, 'P(B|A)', fontsize=12, color=text_color_ab, ha='center')

# 条件付き確率 P(A|B) の説明（BからAへの矢印）
arrow_connection2 = ConnectionPatch((1.5, 0.5), (-1.5, 0.5), 'data', 'data',
                                   arrowstyle='->', mutation_scale=15,
                                   linewidth=2, color=text_color_ab)
ax.add_patch(arrow_connection2)
ax.text(0, 0.8, 'P(A|B)', fontsize=12, color=text_color_ab, ha='center')

# ベイズの定理の応用例を追加
example_title = ax.text(0, -5.2, 'ベイズの定理の応用例', fontsize=16, fontweight='bold', ha='center')

examples = [
    '・医療診断：検査陽性という結果(B)から病気である確率(A|B)を推定',
    '・スパムフィルタ：メッセージ内容(B)からスパムである確率(A|B)を計算',
    '・機械学習：学習データ(B)から最適なモデル(A)の確率を推定'
]

for i, ex in enumerate(examples):
    ax.text(-4.4, -5.7 - i*0.5, ex, fontsize=12)

# レイアウト調整
plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.05)

# 保存と表示
plt.savefig('bayes_theorem_diagram.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.tight_layout()
plt.show()
```
